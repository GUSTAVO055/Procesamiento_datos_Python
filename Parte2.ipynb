```python
#Sumamos la cantidad de poblacion total
df[['Population']].sum()
```

```python
#Calculamos la proporcion DE tipo de sangre sobre la población en cada pais

df['Totales_O_Positivo'] = df['Population'] * df['O+'] / 100
df['Totales_A_Positivo'] = df['Population'] * df['A+'] / 100
df['Totales_B_Positivo'] = df['Population'] * df['B+'] / 100
df['Totales_AB_Positivo'] = df['Population'] * df['AB+'] / 100
df['Totales_O_Negativo'] = df['Population'] * df['O-'] / 100
df['Totales_A_Negativo'] = df['Population'] * df['A-'] / 100
df['Totales_B_Negativo'] = df['Population'] * df['B-'] / 100
df['Totales_AB_Negativo'] = df['Population'] * df['AB-'] / 100
```

```python
#Muestra los tres paises con mayores donantes universales
por_tipo_univ = df.sort_values(['O-','Country'],ascending=False)
muestra = por_tipo_univ.nlargest(3,'O-')
salida = muestra[['Country','Population','Totales_O_Negativo']]
salida
```

```python
#Grafica los tres paises con mayores donantes universales
por_tipo_univ = df.sort_values(['O-','Country'],ascending=False)
muestra = por_tipo_univ.nlargest(3,'O-')
salida = muestra[['Country','Population','Totales_O_Negativo']]
salida.plot(kind = 'bar')
```

```python
#Sumamos la cantidad de población consultada
df['Population'].sum()
```

```python
#Pais con mayor población
paises = df.loc[(df.Country != 'World') ]
mayor_poblacion = paises['Population'].max()
paises = df.loc[(df.Population == mayor_poblacion) ]
paises[['Country','Population' ]]
```

```python
#Pais con menor población
paises = df.loc[(df.Country != 'World') ]
menor_poblacion = paises['Population'].min()
paises = df.loc[(df.Population == menor_poblacion) ]
paises[['Country','Population' ]]
```

```python
#Promedio de población por paises
df['Population'].mean() 
```

```python
#Mostrar paises con población superior al promedio
promedio = df['Population'].mean()
poblacion_superior = df[df.Population > promedio]
lista = poblacion_superior[['Country','Population']]
lista.sort_values(['Population'],ascending=False)
```

```python
#Mostrar 10 primeros paises con población inferior al promedio
promedio = df['Population'].mean()
poblacion_superior = df[df.Population <= promedio]
lista = poblacion_superior[['Country','Population']]
lista.sort_values(['Population'],ascending=False).head(10)
```

```python
#Mostrar 10 últimos paises con población inferior al promedio
promedio = df['Population'].mean()
poblacion_superior = df[df.Population <= promedio]
lista = poblacion_superior[['Country','Population']]
lista.sort_values(['Population'],ascending=False).tail(10)
```

```python
#Pais con mayor número de donantes universales
paises = df.loc[(df.Country != 'World') ]
maximo_universales = paises['Totales_O_Negativo'].max()
paises = df.loc[(df.Totales_O_Negativo == maximo_universales) ]
paises[['Country','Population','O-','Totales_O_Negativo' ]]
```
